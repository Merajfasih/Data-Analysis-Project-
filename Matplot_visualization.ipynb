Import Necessory Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Load gas Price data

Load CSV file

```python
gas = pd.read_csv('gas_prices.csv')
gas.head()
```

```python
plt.figure(figsize=(8,5))
plt.title('Gas Prices over Time in In USD',fontsize=20)
plt.plot(gas.Year,gas.USA,'b.-',label = 'USA')
plt.plot(gas.Year,gas.Canada,'g.-',label = 'Canada')
plt.plot(gas.Year,gas['South Korea'],'r.-',label = 'South Korea')
plt.xticks(gas.Year[::3])
plt.xlabel('Years',fontsize=16)
plt.ylabel('US Dollar',fontsize=16)
plt.savefig('test.jpg')
plt.legend()
plt.show()
```

# Load fifa data

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.Club
```

# Histogram

```python
bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall,bins,color = '#994d00')
plt.xticks(bins)
plt.title('Distribution of skills in Fifa 2018',fontsize=20)
plt.xlabel('Number of Players',fontsize=18)
plt.ylabel('Skill Level',fontsize=18)
plt.show()
```

```python
fifa['Preferred Foot']
```

```python
left = fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
colors = ['#5900b3','#b3002d']
labels = ['Left','Right']
plt.title('Foot Preferrence of FIFA Players',fontsize = 16)
plt.pie([left,right],labels=labels,colors=colors,autopct='%.2f %%')
plt.show()
```

```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```

```python
plt.style.use('default')
plt.figure(figsize=(5,8))
barecolona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
rev = fifa.loc[fifa.Club == 'Manchester United']['Overall']
labels = ['FC Barcelona' , 'Real Madrid' , 'Manchester United']
boxes = plt.boxplot([barecolona,madrid,rev],labels=labels)
for box in boxes ['boxes']:
    box.set(color = 'r',linewidth = 2)
plt.title('Professional Scoccer team Comparision')
plt.ylabel('FIFA overall rating')
plt.xlabel('CLUBS')
plt.show()
```
